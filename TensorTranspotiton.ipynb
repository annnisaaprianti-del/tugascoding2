Tensor Transpotition

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[18, 8], [1, 8], [2, 5]])
X
```

```python
#matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[18, 8], [1, 8], [2, 5]])
X_pt
```

```python
#matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[18, 8], [1, 8], [2, 5]])
X_tf
```

```python
tf.transpose(X_tf)
```
