```python
import numpy as np
```

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[18, 81], [1, 8], [25, 15]])
X
```

```python
#hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum
X.sum()
```

```python
import tensorflow as tf
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[18, 81], [1, 8], [25, 15]])
X_tf
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X
X.sum(axis=0)
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X
X.sum(axis=1)
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X_tf
tf.reduce_sum(X_tf, 1)
```

```python
import torch
```

```python
X_pt = torch.tensor([[18, 81], [1, 8], [25, 15]])
X_pt
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X_pt
torch.sum(X_pt)
```

```python
#hitung hasil penjumlahan matrix X berdasarkan row nya
sum(X, 1)
```

```python
#hitung hasil penjumlahan matrix X_pt berdasarkan row nya
torch.sum(X_pt, 1)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan row nya
tf.reduce_sum(X_tf, 1)
```

```python
#hitung hasil penjumlahan matrix X berdasarkan kolom nya
X.sum(0)
```

```python
#hitung hasil penjumlahan matrix X_pt berdasarkan kolom nya
torch.sum(X_pt, 0)
```

```python
#hitung hasil penjumlahan matrix X_tf berdasarkan kolom nya
tf.reduce_sum(X_tf, 0)
```

```python
#hitung nilai minimal dari matrix X, X_pt dan X_tf
min_X = X.min()
min_X_pt = X_pt.min()
min_X_tf = tf.reduce_min(X_tf)
print(min_X, min_X_pt, min_X_tf)
```

```python
#hitung nilai rata-rata dari matrix X, X_pt dan X_tf
max_X = X.max()
max_X_pt = X_pt.max()
max_X_tf = tf.reduce_max(X_tf)
print(max_X, max_X_pt, max_X_tf)
```

```python
#hitung hasil perkalian seluruh elemen matrix X
X.prod()
```

```python
#hitung hasil perkalian seluruh elemen matrix X_pt
torch.prod(X_pt)
```

```python
#hitung hasil perkalian seluruh elemen matrix X_tf
tf.reduce_prod(X_tf)
```
